```python
from fake_useragent import UserAgent
ua = UserAgent()
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import requests
```

```python
header = {'User-Agent':str(ua.chrome)}
```

```python
def crawling(soup):
    table_soup = soup.find("table", attrs={'class': 'datatable12'})
    names = []
    addresses = []
    informs = []
    crawled = {'학교명': names, '주소': addresses, '정보': informs}

    for idx, item in enumerate(table_soup.find_all('tr')):
        if idx == 0:
            continue
        school_name = item.find('a').text
        address = item.find('div', attrs={'style' : "clear:both;float:left;margin-top:10px"}).text
        information = item.find('div', attrs={'style' : "margin-top:10px;color:gray;font-size:14px"}).text.strip()
        names.append(school_name)
        addresses.append(address)
        informs.append(information)
    return pd.DataFrame.from_dict(crawled)
```

```python
i = 1
total_df = pd.DataFrame()
while True:
    url = f"https://www.edufindkorea.com/edu/list.php?pagenum={i}&addcode=&searchkeyword=&orderby=count&hashtag=&schoolgubun=%EC%B4%88%EB%93%B1%ED%95%99%EA%B5%90&gubun=&subject="
    response = requests.get(url, headers=header)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content)
        total_df = pd.concat([total_df, crawling(soup)])
        i += 1
    if i == 11:
        break
```
